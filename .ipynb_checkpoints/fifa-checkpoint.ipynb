```python
# Loading the Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# We can use "index" instead of "keys()" also
fifa['nationality'].value_counts()[0:5].keys()
```

```python
# Plotting bar for categorical values as well as numerical also
plt.figure(figsize=(8,5))
plt.bar(list(fifa['nationality'].value_counts()[0:5].keys()),list(fifa['nationality'].value_counts()[0:5]),color="g")
plt.show()
```

```python
player_salary = fifa[['short_name','wage_eur']]
```

```python
player_salary.head()
```

```python
player_salary = player_salary.sort_values(by=['wage_eur'],ascending=False)
```

```python
player_salary.head()
```

```python
plt.figure(figsize=(10,5))
plt.bar(list(player_salary['short_name'])[0:5],list(player_salary['wage_eur'])[0:5],color=['red','blue','green','pink','orange'])
plt.xlabel('Players Name')
plt.ylabel('Players Salary')
plt.show()
```
